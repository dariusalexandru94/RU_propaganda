## Data collection

For data collection, I have used web scraping instead of a Twitter API. I chose this method because of Twitter Standard API limitations in terms of data cantity that can be extracted. The maximum number of requests that are allowed is based on a time interval, some specified period or window of time.

API limiting, which is also known as rate limiting, is an essential component of Internet security, as DoS attacks can tank a server with unlimited API requests.

For collecting the data, I have used the following script which I ran once a day from the beginning of the conflict in Ukraine. A new CSV file was created everyday, with tweets that were posted on that specific day by the chosen accounts.

```python
import twint
import nest_asyncio
nest_asyncio.apply()
import time

c = twint.Config()

s_list = ['RosettaNews247', 'RT_com', 'LumpyLouish', 'Levi_godman', 'RWApodcast', 'Taurevanime', 'realGonzaloLira', 
          'Russ_Warrior', 'SNMilitary', 'GeromanAT', 'thesiriusreport', 'RussianEmbassy', 'gbazov', 'colonelhomsi', 'Vick_top55', 
          'PelmeniPusha', 'Alexxa1721', 'jacksonhinklle', 'GarlandNixon', 'AngieSkys', 'NinaByzantina', 'Tinkzorg', 
          'mfa_russia', 'RussiaUN', 'A__Alimov', 'mission_russian', 'rusembusa', 'StalinFrog']

for name in s_list:
    c.Username = name
    c.Since = '2022-05-09 00:00:00'
    c.Until = '2022-05-09 12:00:00'
    c.Since = '2022-05-09 12:01:00'
    c.Until = '2022-05-09 23:59:00'
    time.sleep(3)
    twint.run.Search(c)
    time.sleep(3)
    twint.run.Search(c)
    time.sleep(3)
    twint.run.Search(c)
    c.Store_csv = True
    c.Output = r"C:\Users\Asus\Desktop\W\rf\rs.csv"
```
